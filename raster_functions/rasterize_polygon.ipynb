```python
import geopandas as gpd
from geocube.api.core import make_geocube
import matplotlib.pyplot as plt
import rasterio
```

```python
vectordata_path= 'C:/Users/xbacos/OneDrive - University of Gothenburg/Artikel_2/Enkat/PostCode/PostCodeGBG.shp'
vectordata = gpd.read_file(vectordata_path)

raster_out_path = ''

# what column to rasterize
attribute = 'POSTALCODE'

# make sure attribute is int or float
vectordata = vectordata.astype({attribute:'int'})
# cell size of raster
cell_size = 10

# rasterize 
cube = make_geocube(
    vector_data = vectordata,
    measurements = [attribute],
    resolution=(-cell_size, cell_size),
)

# plot
cube[attribute].plot(cmap = 'rainbow')

# Save raster
cube[attribute].rio.to_raster(raster_out_path)
```
